Plots from VAE models

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
```

```python
# load datasets - test batch size and latent dimension size
loss1_1 = np.load('loss_VAE1_1.npy')
loss1_2 = np.load('loss_VAE1_2.npy')
loss1_3 = np.load('loss_VAE1_3.npy')

loss1_4 = np.load('loss_VAE1_4.npy')
loss1_5 = np.load('loss_VAE1_5.npy')
loss1_6 = np.load('loss_VAE1_6.npy')

loss1_7 = np.load('loss_VAE1_7.npy')
loss1_8 = np.load('loss_VAE1_8.npy')
loss1_9 = np.load('loss_VAE1_9.npy')
```

```python
# plot VAE with different batch size and latent dimensions

epoch = list(range(1, 16))

plt.plot(epoch, loss1_1, label='B16_L16', color='red', linestyle='--')
plt.plot(epoch, loss1_2, label='B16_L32', color='red', linestyle='-')
plt.plot(epoch, loss1_3, label='B16_L64', color='red', linestyle='-.')
plt.plot(epoch, loss1_4, label='B32_L16', color='purple', linestyle='--')
plt.plot(epoch, loss1_5, label='B32_L32', color='purple', linestyle='-')
plt.plot(epoch, loss1_6, label='B32_L64', color='purple', linestyle='-.')
plt.plot(epoch, loss1_7, label='B64_L16', color='green', linestyle='--')
plt.plot(epoch, loss1_8, label='B64_L32', color='green', linestyle='-')
plt.plot(epoch, loss1_9, label='B64_L64', color='green', linestyle='-.')


plt.xlabel('Epoch')
plt.ylabel('Variational Lower Bound')

plt.title('VAE - different batch size and latent dimension')

plt.legend()

plt.show()
```

```python
# load data - test regularization

loss1_10 = np.load('loss_VAE1_10.npy')
loss1_11 = np.load('loss_VAE1_11.npy')
loss1_12 = np.load('loss_VAE1_12.npy')
loss1_13 = np.load('loss_VAE1_13.npy')
loss1_14 = np.load('loss_VAE1_14.npy')
```

```python
epoch = list(range(1, 16))

# VAE1 plot regularization. batch size 16 and latent dimension 64
plt.plot(epoch, loss1_10, label='SpatialDropout(0.2)', color='purple')
plt.plot(epoch, loss1_11, label='SpatialDropout(0.4)', color='black')
plt.plot(epoch, loss1_12, label='Batchnormalization + SpatialDropout(0.2)', color='green')
plt.plot(epoch, loss1_13, label='BatchNormalization + SpatialDropout(0.4)', color='blue')
plt.plot(epoch, loss1_14, label='BatchNormalization', color='red')
plt.plot(epoch, loss1_3, label='No regularization', color='orange')


plt.xlabel('Epoch')
plt.ylabel('Variational Lower Bound')
plt.title('VAE1 - regularization')
plt.legend()
plt.show()
```
